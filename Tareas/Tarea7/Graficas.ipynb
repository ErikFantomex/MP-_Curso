```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df1 = pd.read_csv("tiempos1.dat", sep=',',  names=['np','tiempo']) #Regular sampling 
df2 = pd.read_csv("tiempos2.dat", sep=',',  names=['np','tiempo']) # Merge split 
df3 = pd.read_csv("numpydatos1.dat", sep='	',  names=["np",'tiempo']) # Merge split lectura paralela 
```

# Grafica Velocidad 

```python
plt.plot(1, 2.675697,color='r',marker='*',markersize=20)
plt.text(1.5, 2.675697,'<--- qsort secuencial',color='r')

plt.plot(df2['np'], df2['tiempo'],color='b',marker='.',label='Mergesort paralelo',linestyle='dashed')
plt.plot(df1['np'], df1['tiempo'],color='g',marker='.',label='Regular sampling')
plt.plot(df3['np'], df3['tiempo'],color='r',marker='.',label='Merge split lectura paralela ')

plt.xlabel("Numero de procesos")


plt.ylabel("Tiempo")
plt.legend()
plt.title(" Merge split lectura paralela  vs Merge-split Sort vs Regular sampling")
#plt.axis([0.5, 10.5, -0.1, maximo+0.5])
```

# Speed up

```python
plt.plot(df2['np'], 2.675697/df2['tiempo'],color='b',marker='.',label='Mergesort paralelo',linestyle='dashed')
plt.plot(df1['np'], 2.675697/df1['tiempo'],color='g',marker='.',label='Regular sampling')
plt.plot(df3['np'],  2.675697/df3['tiempo'],color='r',marker='.',label='Merge split lectura paralela ')

plt.xlabel("Numero de procesos")

plt.ylabel("Speedup")
plt.legend()
plt.title(" Merge-split Sort vs Regular sampling vs Merge split lectura paralela ")
#plt.axis([0.5, 10.5, -0.1, maximo+0.5])
```

# Eficiencia

```python
plt.plot(df2['np'], (2.675697/df2['tiempo'])*df2['np'],color='b',marker='.',label='Mergesort paralelo',linestyle='dashed')
plt.plot(df1['np'], (2.675697/df1['tiempo'])*df1['np'],color='g',marker='.',label='Regular sampling')
plt.plot(df3['np'],  (2.675697/df3['tiempo'])*df3['np'],color='r',marker='.',label='Merge split lectura paralela ')

plt.xlabel("Numero de procesos")
plt.ylabel("Eficiencia")
plt.legend()
plt.title(" Merge-split Sort vs Regular sampling vs Merge split lectura paralela")
#plt.axis([0.5, 10.5, -0.1, maximo+0.5])
```

# costo

```python
plt.plot(df2['np'], (2.675697/df2['tiempo'])*(df2['np']*df2['tiempo']),color='b',marker='.',label='Mergesort paralelo',linestyle='dashed')
plt.plot(df1['np'], (2.675697/df1['tiempo'])*(df1['np']*df1['tiempo']),color='g',marker='.',label='Regular sampling')
plt.plot(df3['np'],  (2.675697/df3['tiempo'])*(df3['np']*df3['np']),color='r',marker='.',label='Merge split lectura paralela ')

plt.xlabel("Numero de procesos")
plt.ylabel("Costo")
plt.legend()
plt.title(" Merge-split Sort vs Regular sampling vs Merge split lectura paralela ")
#plt.axis([0.5, 10.5, -0.1, maximo+0.5])
```
